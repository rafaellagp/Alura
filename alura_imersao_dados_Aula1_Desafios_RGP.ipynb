```python
#desafio 1.Realizar a média da metragem para cara um dos bairros (Paulo)
```

```python
dados["Metragem"].mean() #media da metragem de todos os bairros
```

```python
bairros = dados['Bairro'].values # Nome dos bairros.
```

```python
bairros
```

```python
dict_media_metragem = {}

for bairro in bairros:
  media_metragem = dados.loc[dados['Bairro'] == bairro, 'Metragem'].mean() # Calcula a média da metragem de acordo com o nome do bairro.
  dict_media_metragem[bairro] = media_metragem # Adiciona a média ao dicionário.
```

```python
dict_media_metragem
```

```python
#Indo além! Fazer uma média dos valores dos imoveis de cada bairro. Para isso é necessario converter 'Valor' de object para int64
```

```python
dados.dtypes
```

```python
dados = dados[dados['Valor'].str.contains('/Ano')==False]
dados = dados[dados['Valor'].str.contains('/Dia')==False]
```

```python
dados['Valor'] = dados['Valor'].str.replace('/Mês','')
dados['Valor'] = dados['Valor'].str.replace('\n','')
dados['Valor'] = dados['Valor'].str.slice(2)
dados['Valor'] = dados['Valor'].str.replace(' ','')
dados['Valor'] = dados['Valor'].str.replace('.','')

```

```python
dados['Valor'] = pd.to_numeric(dados['Valor'])
```

```python
dados.dtypes
```

```python
bairros = dados['Bairro'].values # Nome dos bairros.
```

```python
dict_media_valores = {}

for bairro in bairros:
  media_valores = dados.loc[dados['Bairro'] == bairro, 'Valor'].mean() # Calcula a média dos valores de acordo com o nome do bairro.
  dict_media_valores[bairro] = media_valores # Adiciona a média ao dicionário.
```

```python
#Desafio 4 - Pegar outros estatísticas dos dados, média, mediana, mim, max (Vivi)
```

```python
print(dados['Valor'].mean()) #Média Qual a média de preços de imoveis em São Paulo?
```

```python
print(dados['Valor'].median()) #Mediana Qual o valor médio de um imovel em São Paulo?
```

```python
print(dados['Valor'].quantile()) #quartil Qual o valor médio de um imovel em São Paulo?
```

```python
print(dados['Valor'].mode()) #moda Qual o valor mais comum de imoveis em São Paulo?
```

```python
print(dados['Valor'].min()) #Qual o valor do imovel mais barato em São Paulo?
```

```python
print(dados['Valor'].max()) #Qual o valor do imovel mais caro em São Paulo?
```

```python
print(dados['Valor'].std()) #Qual a variação de preços em relação ao preço médio dos imoveis de São Paulo? (desvio padrão)
```

```python
# Média
mdDados = dados.Metragem.mean()
# Mediana
mdnDados = dados.Metragem.median()
# Mínima
minDados = dados.Metragem.min()
# Máxima
maxDados = dados.Metragem.max()
```
